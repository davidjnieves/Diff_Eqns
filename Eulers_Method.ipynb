# Numerical Approximations: Euler's Method 
We will Consider the initial value problem (IVP) $$y' = 2(t+1)y,\quad y(0) = 1.$$
The cells below contain python code that implements Euler's Method. Modify this code to solve the above IVP. Note: as written, the code approximates the solution to the IVP $$y' =-2ty + t,\quad y(0) = -1.$$.

To evalute cells, type shift + enter.

```python
import numpy as np
from matplotlib import pyplot as plt
from ipywidgets import interactive, fixed
```

```python
# Routine for Euler's Method to Solve Initial Value Problems
# Input: interval endpoints [a,b]: a, b
#        initial value: y0
#        number of steps: n
# Output: time steps t, approximate solution y
def euler1(a,b,y0,n):

    # initialize variables
    # n steps to be performed plus initial condition: n+1
    t = np.zeros(n+1)
    y = np.zeros(n+1)

    # Apply initial conditions
    t[0] = a
    y[0] = y0

    # Step size
    h = (b-a)/n

    # Right-hand-sdie of differential equation
    def f(t,y):
        y_prime = -2*t*y + t
        return y_prime

    for i in range(n):
        # advance time from t_i to t_(i+1)
        t[i+1] = t[i] + h
        # Euler's method for the approximation at time t_(i+1) 
        y[i+1] = y[i] + h*f(t[i],y[i])

    # plot data
    fig, ax = plt.subplots()       # Create a figure containing a single Axes.
    ax.plot(t,y,marker = 'o')      # Plot some data on the Axes.
    plt.show()                     # Show the figure.
    
    #variables to be returned
    return t, y
```

```python
euler1(0,0.4,-1,4)
```
