# Numerical Approximations: Euler's Method 
We will Consider the initial value problem
	(IVP) $$y' = 2(t+1)y,\quad y(0) = 1.$$

```python
import numpy as np
from matplotlib import pyplot as plt
from ipywidgets import interactive, fixed
```

```python
# Euler's Method for Solving Initial Value Problems
# Input: interval endpoints [a,b], 
#        initial value y0, 
#        number of steps n
# Output: time steps t, solution y
def euler1(a,b,y0,n):

    # initialize variables
    t = np.zeros(n)
    y = np.zeros(n)

    # Apply initial conditions
    t[0] = a
    y[0] = y0

    # Step size
    h = (b-a)/n

    # Right-hand-sdie of differential equation
    def f(t,y):
        y_prime = -10*y
        return y_prime

    for i in range(1,n):
        # advance time from t_i to t_(i+1)
        t[i] = t[i-1] + h
        
        # Euler's method for the approximation at time t_(i+1) 
        y[i] = y[i-1] + h*f(t[i-1],y[i-1])

    # plot data
    fig, ax = plt.subplots()       # Create a figure containing a single Axes.
    ax.plot(t,y,marker = 'o')      # Plot some data on the Axes.
    plt.show()                     # Show the figure.
    
    #variables to be returned
    return #t, y
```

```python
euler1(0,1,1,25)
```
